### Estrutura de repetição

Imagine que uma estrutura de repetição é como se fosse um robô que faz tarefas repetidas pra você. Ele recebe uma lista de coisas e sabe que precisa fazer a mesma coisa sempre para cada item.

Sem o `for`, nossa estrutura de repetição utilizada, se você quisesse pegar cada item da sua lista de compras e mostrar na tela, você teria que dar uma ordem para cada um, como se estivesse dizendo:

"Robô, mostre o primeiro item da lista! Agora mostre o segundo! E o terceiro... e o quarto..."

```python
#Fazer print de cada item da lista

lista_compras = ['tênis','ovos','ventilador','caneta','refrigerante'] #Essa é nossa lista de compras
print(lista_compras[0])
print(lista_compras[1])
print(lista_compras[2])
print(lista_compras[3])
print(lista_compras[4])
```

É aqui que o nosso robô entra. Com o `for`, você pode dar uma única ordem para ele:

"Robô, para cada item nesta lista, mostre o item na tela."

```python
# Para CADA ITEM na nossa lista de compras, faça o seguinte:
for item in lista_compras:
    print(item)
```

### ZIP

Imagine que você tem duas listas, uma com nomes e outra com idades dos seus pacientes, e quer mostrar esses dados juntos.

O `zip` pode ajudar nisso, ele pega a primeira coisa de uma lista, e a primeira coisa da outra lista.

O processo abaixo não usa o `zip`.

```python
nomes = ['João', 'Maria', 'Pedro']
idades = [35, 42, 28]

# Fazemos a combinação "na mão", um por um.
print(f"Nome: {nomes[0]}, Idade: {idades[0]}")
print(f"Nome: {nomes[1]}, Idade: {idades[1]}")
print(f"Nome: {nomes[2]}, Idade: {idades[2]}")
```

Já com ele, ficaria assim

```python
nomes = ['João', 'Maria', 'Pedro']
idades = [35, 42, 28]

# O comando zip() vai parear as duas listas para nós.
# Em cada rodada, "nome" e "idade" serão o par correspondente.
for nome, idade in zip(nomes, idades):
    # O computador já sabe quem é o par de quem!
    print(f"Nome: {nome}, Idade: {idade}")
```

## O que é um DataFrame?

Um DataFrame é como se fosse uma tabela de Excel que você usa dentro do Colab.

Ele é uma forma super organizada de guardar dados em linhas e colunas. Cada coluna tem um tipo de informação por exemplo, `Nome`, `Numero de Cáries`, `Custo do Tratamento (R$)` e cada linha representa um "registro" por exemplo, os dados de um paciente.

```python
# Primeiro, vamos "chamar" uma ferramenta especial para criar tabelas.
# Essa ferramenta se chama 'pandas' e vamos usar a sigla 'pd' para ficar mais fácil de usar.
import pandas as pd

#Agora, vamos preencher nossa tabela com alguns dados.
dados_pacientes = {
    "Nome": ["João", "Maria", "Pedro", "Ana", "Lucas"],
    "Número de Cáries": [1, 2, 4, 6, 8],
    "Custo do Tratamento (R$)": [150.00, 250.00, 500.00, 700.00, 950.00]
}

# Agora, vamos transformar os dados em um DataFrame.
# Usamos a ferramenta 'pd' para criar um "DataFrame" (que é o nome chique para a tabela).
# E guardamos essa tabela na variável 'dados_pacientes'.
dados_pacientes = pd.DataFrame(dados_pacientes)

# Por fim, vamos pedir para o computador "mostrar" o que tem dentro da nossa tabela 'dados_pacientes'.
dados_pacientes
```

## O que é Plotar os Dados?

Imagine que você tem a sua tabela (o nosso DataFrame) cheia de números e informações dos pacientes. Ler esses números um por um pode ser confuso e cansativo...

Plotar os dados é como transformar os dados da tabela em um desenho. Em vez de ver uma lista de números, você vai ver um gráfico, o que torna muito mais fácil entender o que está acontecendo com os dados.

## **2)** Substitua os valores dos dados da sua tabela nos modelos A, B e C, para prever o comportamento dos dados nos dias seguintes. Vamos ver com o exemplo abaixo como fazer este procedimento passo a passo.

### Vamos começar fazendo a previsão para o modelo A, Grupo 1.

#### Com a tabela inicial, fazemos a previsão do Dia 11/04.

```python
# Dados do Grupo 1 (Limpezas).
# Vamos pegar os últimos 3 e 2 valores da coluna "Limpezas" para fazer o cálculo.
# Lembre-se que o Colab começa a contar do 0!
# [5, 6, 7, 6, 8, 9, 10, 10, 11, 13]
# Os últimos 3 valores são 10, 11 e 13.
# Os últimos 2 valores são 11 e 13.

# Primeiro, vamos guardar os últimos 3 valores em uma variável
ultimos_3_valores = df_limpezas["Limpezas"].tail(3)

# Agora, vamos guardar os últimos 2 valores em outra variável
ultimos_2_valores = df_limpezas["Limpezas"].tail(2)

# O Modelo A usa a soma dos últimos 3 valores e a média dos últimos 2
soma_3 = ultimos_3_valores.sum()
media_2 = ultimos_2_valores.mean()

# E o resultado é a nossa previsão para o futuro!
previsao_1 = ((0.6 * soma_3) + (0.4 * media_2)) / 3

print(f"Previsão do Modelo A para o Grupo de Limpezas dia 11/04: {previsao_1:.2f}")
```

Agora, vamos criar uma nova tabela (DataFrame) de uma linha só para a nossa previsão.

```python
df_previsao_1 = pd.DataFrame({"Data": [pd.to_datetime("2025-04-11")], "Limpezas": [previsao_1]})

# Vamos ver como ficou essa nova tabela
print("Tabela da primeira previsão:")
df_previsao_1
```

Usamos o comando pd.concat() para juntar as duas tabelas.

Pense no `pd.concat()` como um "colar" duas tabelas do Excel uma embaixo da outra.

Isso permite usar o novo valor para o próximo cálculo.

```python
# O ignore_index=True serve para ele organizar os números das linhas direitinho.
df_limpezas_previsao_1_modelo_a = pd.concat([df_limpezas, df_previsao_1], ignore_index=True)

# Agora, nossa tabela atualizada tem a primeira previsão no final!
df_limpezas_previsao_1_modelo_a
```

#### Com a tabela atualizada, fazemos a previsão do Dia 12/04.

Lembra do processo que fizemos passo a passo da previsão_1?

Agora ele está sendo feito aqui, porém tudo em uma só linha, seguindo exatamente a fórmula!

```python
previsao_2 = ((0.6 * df_limpezas_previsao_1_modelo_a["Limpezas"].tail(3).sum()) + (0.4 * df_limpezas_previsao_1_modelo_a["Limpezas"].tail(2).mean())) / 3

print(f"Previsão do Modelo A para o Grupo de Limpezas dia 12/04: {previsao_2:.2f}")
```

```python
# Criamos a tabela da segunda previsão
df_previsao_2 = pd.DataFrame({"Data": [pd.to_datetime("2025-04-12")], "Limpezas": [previsao_2]})

# Juntamos as tabelas novamente para ter a mais nova em mãos
df_limpezas_previsao_2_modelo_a = pd.concat([df_limpezas_previsao_1_modelo_a, df_previsao_2], ignore_index=True)

# Agora, nossa tabela atualizada tem a segunda previsão no final!
df_limpezas_previsao_2_modelo_a
```

#### Com a tabela atualizada, fazemos a previsão do Dia 13/04.

```python
# Fazemos a conta usando a tabela com a previsão do dia 11/4 e 12/4
previsao_3 = ((0.6 * df_limpezas_previsao_2_modelo_a["Limpezas"].tail(3).sum()) + (0.4 * df_limpezas_previsao_2_modelo_a["Limpezas"].tail(2).mean())) / 3

# Mostramos o resultado da previsão
print(f"Previsão do Modelo A para o Grupo de Limpezas dia 13/04: {previsao_3:.2f}")

# Criamos a tabela da segunda previsão
df_previsao_3 = pd.DataFrame({"Data": [pd.to_datetime("2025-04-13")], "Limpezas": [previsao_3]})

# Juntamos as tabelas novamente para ter a mais nova em mãos
df_limpezas_previsao_3_modelo_a_g1 = pd.concat([df_limpezas_previsao_2_modelo_a, df_previsao_3], ignore_index=True)

# Agora, nossa tabela atualizada tem a última previsão no final!
df_limpezas_previsao_3_modelo_a_g1
```

Dessa forma, fizemos a previsão dos dados do grupo 1 para o modelo A. Mas, como já vimos, existem as funções, que são muito úteis nessas horas.

Você percebeu que repetimos alguns processos durante a resolução, então podemos transformar isso em uma função!

```python
def prever_proximo_dia_modelo_a(tabela_de_dados,tipo):
    # A função faz o cálculo do Modelo A.
    soma_ultimos_3 = tabela_de_dados[tipo].tail(3).sum()
    media_ultimos_2 = tabela_de_dados[tipo].tail(2).mean()

    # O resultado final é a nossa previsão
    previsao = ((0.6 * soma_ultimos_3) + (0.4 * media_ultimos_2)) / 3

    # A função nos devolve o valor da previsão.
    return previsao
```

Vamos testar essa função

```python
previsao_1_com_funcao=prever_proximo_dia_modelo_a(df_limpezas,"Limpezas")
print(previsao_1_com_funcao)
```

Olhe só! No dia 11/04 o valor previsto é exatamente o encontrado por nossa função... Porem, devemos ficar chamando essa função toda hora?

A resposta é não. Existe outro comando que pode nos auxiliar ainda mais nesse processo, o `for`, com ele vai ser possível prever todos os dias de maneira muito prática. Vamos lá!

```python
# Vamos fazer a previsão para os próximos 3 dias
dias_para_prever = 3

# Usamos uma cópia da tabela original para não mexer nela.
df_limpezas_previsao_3_modelo_a_g1 = df_limpezas_org.copy()

# O "for" vai chamar a nossa função 3 vezes, uma para cada dia.
for i in range(dias_para_prever):
    # 1. Chamamos a nossa função para nos dar a previsão do próximo dia.
    previsao_calculada = prever_proximo_dia_modelo_a(df_limpezas_previsao_3_modelo_a_g1,"Limpezas")

    # 2. Criamos uma nova linha com o resultado.
    ultima_data = df_limpezas_previsao_3_modelo_a_g1["Data"].iloc[-1]
    nova_linha = pd.DataFrame({"Data": [ultima_data + pd.Timedelta(days=1)], "Limpezas": [previsao_calculada]})

    # 3. Adicionamos a nova linha à tabela.
    df_limpezas_previsao_3_modelo_a_g1 = pd.concat([df_limpezas_previsao_3_modelo_a_g1, nova_linha], ignore_index=True)
```

Agora todo aquele processo que repetimos uma vez para cada dia foi feito somente nessa célula de código. Olhe o resultado!

### Agora, vamos fazer a previsão do modelo B, para o Grupo 1.

Como já aprendemos que podemos utilizar a função e o for para auxiliar na construção da nossa solução, vamos continuar usando essa abordagem aqui!

```python
def prever_proximo_dia_modelo_b(tabela_de_dados,tipo):
    # Dentro da função, ela faz o cálculo do Modelo B.
    # 1. Pega o último valor da nossa tabela
    ultimo_valor = tabela_de_dados[tipo].iloc[-1]

    # 2. Calculamos a média das diferenças dos últimos 4 dias
    # O comando .diff() calcula a diferença de um dia para o outro.
    # O comando .tail(3) pega as 3 últimas diferenças.
    media_diferencas = tabela_de_dados[tipo].diff().tail(3).mean()

    # 3. Faz o cálculo final da previsão, conforme a fórmula do Modelo B.
    previsao = ultimo_valor + media_diferencas

    # A função nos "devolve" o valor da previsão.
    return previsao
```

Agora que definimos a função para calcular o modelo B só precisamos chamar ela no for com o dataframe correto.

```python
dias_para_prever = 3

# Usamos uma cópia da tabela original para não mexer nela.
df_limpezas_previsao_3_modelo_b_g1 = df_limpezas_org.copy()

# O "for" vai chamar a nossa função 3 vezes, uma para cada dia.
for i in range(dias_para_prever):
    # 1. Chamamos a nossa função para nos dar a previsão do próximo dia.
    previsao_calculada = prever_proximo_dia_modelo_b(df_limpezas_previsao_3_modelo_b_g1,"Limpezas")

    # 2. Criamos uma nova linha com o resultado.
    ultima_data = df_limpezas_previsao_3_modelo_b_g1["Data"].iloc[-1]
    nova_linha = pd.DataFrame({"Data": [ultima_data + pd.Timedelta(days=1)], "Limpezas": [previsao_calculada]})

    # 3. Adicionamos a nova linha à tabela.
    df_limpezas_previsao_3_modelo_b_g1 = pd.concat([df_limpezas_previsao_3_modelo_b_g1, nova_linha], ignore_index=True)

df_limpezas_previsao_3_modelo_b_g1
```

Você percebeu que o processo do for é o mesmo feito anteriormente para o modelo A?

Então nos também podemos fazer uma função para ajudar nesse processo, olhe

```python
# Ela recebe a tabela, a "receita" do modelo e a quantidade de dias para prever e o tipo de procedimento.
def prever_dias_sequenciais(tabela_original, funcao_modelo, dias_para_prever, tipo):
    tabela_completa = tabela_original.copy() # Copiamos a tabela original

    # O for loop vai repetir o processo o número de vezes que pedimos
    for i in range(dias_para_prever):
        # A MÁGICA: Chamamos a função do modelo que foi passada!
        previsao_calculada = funcao_modelo(tabela_completa,tipo)

        # Criamos e adicionamos a nova linha à tabela
        ultima_data = tabela_completa["Data"].iloc[-1]
        nova_linha = pd.DataFrame({"Data": [ultima_data + pd.Timedelta(days=1)], tipo: [previsao_calculada]})
        tabela_completa = pd.concat([tabela_completa, nova_linha], ignore_index=True)

    return tabela_completa
```

### Fazendo a previsão do modelo C, para o Grupo 1

Vamos testar essa função com a nossa função do modelo C e com os dados do grupo 1.

```python
#Primeiro definimos a nossa função do modelo C que ainda não tinha sido feita.
def prever_proximo_dia_modelo_c(tabela_de_dados,tipo):
    # Pegamos os últimos 3 valores da tabela e tiramos a média.
    previsao = tabela_de_dados[tipo].tail(3).mean()
    return previsao
```

Agora basta chamar a função principal para que tudo funciona corretamente.

## **3)** Agora que você já tem os dados de previsão calculados, iremos calcular o erro entre eles e os dados reais, e também a média desses erros. Vamos ver com o exemplo abaixo como fazer este procedimento passo a passo.

### Vamos fazer os cálculos de erro para o Grupo 1

 A nossa lista com os valores que realmente aconteceram.

```python
# Pense nisso como a "lista de valores reais".
valores_reais = [14, 15, 14]
```

A nossa lista com os valores que o Modelo A previu.

```python
# Pense nisso como a "lista da previsão".
valores_previstos = df_limpezas_previsao_3_modelo_a_g1["Limpezas"].tail(dias_para_prever).tolist()
```

Passo 1: Calcular o erro individual (Real - Previsto)

Aqui, o computador passa por cada dia e subtrai a previsão do valor real.

O "zip" é um comando que junta as duas listas, como se estivesse pareando.

O valor real com o previsto para que a conta seja feita para cada dia.

```python
erros = [real - previsto for real, previsto in zip(valores_reais, valores_previstos)]
```

Passo 2: Calcular o erro absoluto

O comando "abs()" transforma os números negativos em positivos.
Isso nos mostra a "distância" do erro, sem se importar se a previsão foi para mais ou para menos.

```python
erros_absolutos_modelo_a_g1 = [abs(erro) for erro in erros]
```

Passo 3: Calcular a média do erro absoluto

Agora, o computador soma todos os erros absolutos e divide pela quantidade de erros, para nos dar a média.

```python
media_erro_absoluto_a_g1 = sum(erros_absolutos_modelo_a_g1) / len(erros_absolutos_modelo_a_g1)
```

Passo 4: mostrar os resultados finais

```python
# Pedimos para o computador nos mostrar o resultado de cada etapa do cálculo.
print("Resultados do cálculo do erro (Modelo A - Grupo 1):")
print(f"Valores Reais:              {valores_reais}")
print(f"Valores Previstos:          {[f'{p:.2f}' for p in valores_previstos]}")
print(f"Erros (Real - Previsto):      {[f'{e:.2f}' for e in erros]}")
print(f"Erros Absolutos:            {[f'{a:.2f}' for a in erros_absolutos_modelo_a_g1]}")
print(f"Média Absoluta do Erro: {media_erro_absoluto_a_g1:.2f}")
```

Do mesmo jeito que na questão dois, temos um processo que será repetido para os grupos, então vamos fazer uma função para calcular o erro.

```python
# A nossa função para calcular a Média Absoluta do Erro (MAE).
# Ela precisa de duas listas: uma com os valores reais e outra com os previstos.
def calcular_mae(valores_reais, valores_previstos):
    # --- PASSO 1: CALCULAR O ERRO INDIVIDUAL (Real - Previsto) ---
    erros = [real - previsto for real, previsto in zip(valores_reais, valores_previstos)]

    # --- PASSO 2: CALCULAR O ERRO ABSOLUTO ---
    erros_absolutos = [abs(erro) for erro in erros]

    # --- PASSO 3: CALCULAR A NOTA FINAL: A MÉDIA ---
    media_erro_absoluto = sum(erros_absolutos) / len(erros_absolutos)

    print(f"Valores Reais:              {valores_reais}")
    print(f"Valores Previstos:          {[f'{p:.2f}' for p in valores_previstos]}")
    print(f"Erros (Real - Previsto):      {[f'{e:.2f}' for e in erros]}")
    print(f"Erros Absolutos:            {[f'{a:.2f}' for a in erros_absolutos]}")
    print(f"Média Absoluta do Erro: {media_erro_absoluto:.2f}")
    # A função nos devolve o resultado final.

    return media_erro_absoluto
```

Agora que temos nossa função de cálculo definida, vamos usar para os próximos casos.

```python
# Pense nisso como a "lista de valores reais".
valores_reais = [14, 15, 14]

# A nossa lista com os valores que o Modelo B previu.
valores_previstos = df_limpezas_previsao_3_modelo_b_g1["Limpezas"].tail(dias_para_prever).tolist()

print("Resultados do cálculo do erro (Modelo B - Grupo 1):")
media_erro_absoluto_b_g1 = calcular_mae(valores_reais, valores_previstos)
```

Vamos fazer a função que gera o gráfico levando em consideração os valores reais, os valores previstos e o erro.

```python
def gerar_grafico_comparacao(eixo, df_original, df_previsao_completo, df_validacao, titulo, cor_previsao, tipo, erro_mae):
    # Plotamos os dados originais
    eixo.plot(df_original["Data"], df_original[tipo], marker="o", color="blue", label="Dados Originais")

    # Plotamos a linha da previsão
    eixo.plot(df_previsao_completo["Data"], df_previsao_completo[tipo], linestyle="--", marker="x", color=cor_previsao, label=f"Previsão {titulo}")

    # Plotamos os valores reais de validação (as estrelas)
    eixo.plot(df_validacao["Data"], df_validacao[tipo], marker="*", linestyle="", color="black", markersize=12, label="Valores Reais (Validação)")

    # Usamos um "for" para desenhar uma linha para cada dia de previsão.
    for data_prevista, valor_real, valor_previsto in zip(df_validacao["Data"], df_validacao[tipo], df_previsao_completo[tipo].tail(len(df_validacao))):
        # Desenhamos uma linha vertical que vai do valor real ao previsto.
        eixo.plot([data_prevista, data_prevista], [valor_real, valor_previsto], color="red", linestyle="-", linewidth=1, alpha=0.7)

    eixo.set_title(f"{titulo} -- Erro Absoluto: {erro_mae:.2f}")
    eixo.set_ylabel(f"Quantidade de {tipo}")
    eixo.grid(True)
    eixo.legend()
```
