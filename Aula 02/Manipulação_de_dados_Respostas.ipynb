```python
import pandas as pd

# Carregar o arquivo CSV
file_path = "data/Titanic-Dataset.csv"  # Substitua pelo caminho correto do arquivo entre as " "
df = pd.read_csv(file_path)
df.head()
```

## Ver quantidade de linhas e colunas da base



```python
# Pode usar o .shape para ver as dimensões do DataFrame.
# O primeiro valor é o número de linhas e o segundo é o número de colunas.
print(f"O DataFrame tem {df.shape[0]} linhas e {df.shape[1]} colunas.")

# Mais fácil:
print(df.shape)
```

## Ver os Tipos de Dados



```python
# O método .dtypes retorna o tipo de dado de cada coluna.
df.dtypes
```

## Identificar os Dados Nulos

```python
# Lembre-se do .isna() e do .sum() > 0
df.isna().sum() > 0
```

## Fazer Histograma de uma Coluna

```python
import matplotlib.pyplot as plt

df['Age'].hist() # O método .hist() para histogramas. Qual valor irá pra dentro das ''
```

## Filtrar Dados das Colunas



```python
# Cria uma nova variável (ou sobrescreve a existente) com os dados filtrados.
# A sintaxe é 'df[condição]'.
# Aqui, a condição é 'df[' '] > '.
df_filtrado = df[df['Age'] > 40] # Adicione uma coluna para ser filtrada dentro das '', e um valor para ser maior que.
df_filtrado
```

## Calcular Estatística Básica de uma Coluna

```python
# Calcula estatísticas básicas para a coluna ' '?
# .count(): quantidade de valores não nulos.
# .mean(): média.
# .median(): mediana.
# .std(): desvio padrão.
# .mode(): moda.
print("Estatísticas para a coluna 'Age':")
print(f"Quantidade: {df['Age'].count()}")
print(f"Média: {df['Age'].mean():.2f}")
print(f"Mediana: {df['Age'].median()}")
print(f"Desvio Padrão: {df['Age'].std():.2f}")
print(f"Moda: {df['Age'].mode()[0]}")

# Adicione uma coluna dentro das ' '.
```

## Fazer Gráfico de Linhas

```python
# Primeiro agrupo o dataframe com o .groupby, e conta quantas linhas existem em cada grupo
df_pessoas_por_idade = df.groupby("Age")["Age"].count() # vamos agrupor por qual coluna nas primeiras " "? e contar por qual na segunda " "?

# Criando gráfico de linhas
plt.figure(figsize=(8, 5))
plt.plot(df_pessoas_por_idade)
plt.title("Quantidade de Pessoas por Idade")
plt.xlabel("Idade") # Adicione conforme a escolha das colunas
plt.ylabel("Quantidade de Pessoas") # Adicione conforme a escolha das colunas
plt.grid(True)
plt.show()
```

## Como Substituir Dados Nulos

```python
#LEMBRE-SE DA CÓPIA!
df_copia_constante = df.copy()

# Substituir por uma constante
df_copia_constante = df_copia_constante.fillna(0) # Qual o restante do comando mesmo?

print("Valores nulos =",df_copia_constante["Age"].isna().sum()) # Se for zero está certo.
```

```python
#LEMBRE-SE DA CÓPIA!
df_copia_media = df.copy()
# Substituir valores nulos pela média da coluna
# Qual o restante do comando mesmo?
df_copia_media['Age'] = df_copia_media['Age'].fillna(df_copia_media['Age'].mean())  # Qual coluna vamos usar?

print("Valores nulos =",df_copia_media["Age"].isna().sum()) # Se for zero está certo.
```

```python
#LEMBRE-SE DA CÓPIA!
df_copia_mediana = df.copy()
# Substituir valores nulos pela mediana da coluna
# Qual o restante do comando mesmo?
df_copia_mediana['Age'] = df_copia_mediana['Age'].fillna(df_copia_mediana['Age'].median()) # Qual coluna vamos usar?

print("Valores nulos =",df_copia_mediana["Age"].isna().sum()) # Se for zero esta certo.
```
